Task 6: Create a Decision Tree Classifier and visualize it graphically. And if we feed the classifier a new set of data, it should be able to predict the right class.

***

Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
```

***

Getting the dataset

```python
data = pd.read_csv('Iris.csv')
```

```python
#Looking at the top 5  elements
data.head()            
```

***

Let's see it in a plot

```python
sns.relplot(x = 'PetalWidthCm', y = 'PetalLengthCm', data = data, hue = 'Species', palette = ['red', 'teal', 'violet'],
            height = 7, aspect = 1.5, style = data['Species'], size=0, sizes = (69,69))
plt.title('Petal length vs width')
plt.show()
```

***

Splitting the dataset into train and test sets

```python
# Let's seperate the input and output
data_x = data.drop(['Species','Id'], axis = 1)
data_y = data['Species']
```

```python
print('-->Input data:\n',data_x[:5],'\n\n-->Output data:\n',data_y[:5])
```

```python
# Splitting
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size = 0.2, random_state = 42)
```

```python
print('x train size: ',x_train.shape,'\ny train size: ',y_train.shape,'\n\n:> x test size: ',x_test.shape,
      '\n:> y test size: ',y_test.shape)
```

***

Training using Decision tree

```python
# Defining a decision tree variable
dtc = DecisionTreeClassifier()

# Training the decision tree on our training data
train = dtc.fit(x_train, y_train)
```

```python
# Predicting our test cases using our trained model
y_predict = dtc.predict(x_test)
```

```python
# Let's look at few results and compare it with the original data
pd.DataFrame({'Actual type ':y_test[:5],'Predicted type ':y_predict[:5]})
```

***

Let's check the accuracy score

```python
from sklearn.metrics import accuracy_score
print('Accuracy of training data: ',dtc.score(x_train,y_train))
print('Accuracy on test data: ',accuracy_score(y_test,y_predict))
```

As we can see that, accuracy is pretty good.

***

Let's visualize the Decision Tree

```python
from sklearn import tree
plt.subplots(figsize=(50,50),facecolor='lightblue')
tree.plot_tree(train , filled = True)
plt.show()
```

Now you can feed any new test data to this classifer and it would be able to predict the right class accordingly.
